```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

def split_dataset(input_file, output_dir="data", test_size=0.2):
    try:
        # ------------------------------
        # 1️⃣ Ensure output directory exists
        # ------------------------------
        os.makedirs(output_dir, exist_ok=True)

        # ------------------------------
        # 2️⃣ Load dataset
        # ------------------------------
        df = pd.read_csv(input_file)
        df.columns = df.columns.str.strip()  # remove extra spaces

        if 'label' not in df.columns:
            raise ValueError("Dataset must have a 'label' column for stratification.")

        # ------------------------------
        # 3️⃣ Split dataset (train/test)
        # ------------------------------
        train_df, test_df = train_test_split(
            df,
            test_size=test_size,
            random_state=42,
            stratify=df['label']
        )

        # ------------------------------
        # 4️⃣ Save datasets safely
        # ------------------------------
        train_path = os.path.join(output_dir, "train_dataset.csv")
        test_path = os.path.join(output_dir, "test_dataset.csv")

        # Save train
        train_df.to_csv(train_path, index=False)
        print(f"Train dataset saved at: {train_path} (size: {len(train_df)})")

        # Save test
        test_df.to_csv(test_path, index=False)
        print(f"Test dataset saved at: {test_path} (size: {len(test_df)})")

    except PermissionError as e:
        print(f"Permission Error: {e}")
        print("Make sure the file is not open elsewhere and you have write access to the folder.")
    except Exception as e:
        print(f"Error: {e}")

# ------------------------------
# Run the split
# ------------------------------
if __name__ == "__main__":
    split_dataset("New Dataset.csv")
```
